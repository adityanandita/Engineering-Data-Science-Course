# Soal 1
Penjelasan metrik-metrik yang digunakan dalam klasifikasi dan implementasi sederhana perhitungan metrik menggunakan python.

## Jawaban Soal 1

Dalam klasifikasi, terdapat beberapa metrik penting yang perlu diperhatikan, antara lain:
1. **Accuracy**: Prosentase total prediksi yang benar.
$$ \text{Accuracy} = \frac{\text{True Positives (TP)} + \text{True Negatives (TN)}}{\text{TP} + \text{False Positives (FP)} + \text{False Negatives (FN)} + \text{TN}} $$

2. **Precision**: Prosentase prediksi positif yang sebenarnya positif.
$$ \text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}} $$
3. **Recall (Sensitivity):** Prosentase kasus positif yang teridentifikasi dengan benar.
$$ \text{Recall} = \frac{\text{TP}}{\text{TP} + \text{FN}} $$
4. **F1 Score**: Rata-rata harmonik dari Precision dan Recall.
$$ \text{F1 Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}} $$

Dimana:
- **TP (True Positives)** : Jumlah kasus positif yang teridentifikasi dengan benar.
- **TN (True Negatives)** : Jumlah kasus negatif yang teridentifikasi dengan benar.
- **FP (False Positives)**: Jumlah kasus negatif yang salah teridentifikasi sebagai positif.
- **FN (False Negatives)**: Jumlah kasus positif yang salah teridentifikasi sebagai negatif.

Metrik-metrik tersebut berguna untuk menilai performa model klasifikasi, dan setiap metrik memiliki keunggulan dan kekurangannya sendiri. Penggunaannya pun sangat bergantung pada konteks masalahnya.

Referensi: https://medium.com/@pararawendy19/memahami-metrik-pada-pemodelan-klasifikasi-29cd5b738ee7

## Implementasi Soal 1

````python
# Berikut adalah fungsi untuk menghitung akurasi, presisi, recall, dan f1-score
def calculate_metrics(metric, y_true, y_pred):
    # Hitung True Positive (TP), True Negative (TN), False Positive (FP), dan False Negative (FN)
    TP, TN, FP, FN = 0, 0, 0, 0 # Inisialisasi TP, TN, FP, FN dengan 0
    for i in range(len(y_true)):
        if y_true[i] == 1 and y_pred[i] == 1:
            TP += 1
        elif y_true[i] == 0 and y_pred[i] == 0:
            TN += 1
        elif y_true[i] == 0 and y_pred[i] == 1:
            FP += 1
        elif y_true[i] == 1 and y_pred[i] == 0:
            FN += 1

    accuracy = (TP + TN) / (TP + TN + FP + FN)
    precision = TP / (TP + FP) if TP + FP != 0 else 0
    recall = TP / (TP + FN) if TP + FN != 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if precision + recall != 0 else 0

    if metric == 'accuracy':
        return accuracy
    elif metric == 'precision':
        return precision
    elif metric == 'recall':
        return recall
    elif metric == 'f1-score':
        return f1_score
    else:
        return None

# Contoh
y_true = [1, 1, 0, 0, 0, 1, 0, 0, 1, 0]
y_pred = [1, 0, 1, 0, 0, 1, 1, 0, 1, 1]

print(f"Accuracy: {calculate_metrics('accuracy', y_true, y_pred):.2f}")
print(f"Precision: {calculate_metrics('precision', y_true, y_pred):.2f}")
print(f"Recall: {calculate_metrics('recall', y_true, y_pred):.2f}")
print(f"F1-Score: {calculate_metrics('f1-score', y_true, y_pred):.2f}")
````

# Soal 2
Penjelasan `confusion matrix` dan kurva `receiver-operating characteristic` dalam konteks klasifikasi.

## Jawaban Soal 2

$\textit{Confusion matrix}$ merupakan matriks yang menunjukkan jumlah prediksi yang benar dan salah dari model klasifikasi untuk setiap kelas. Pada confusion matrix terdapat baris yang merepresentasikan label sebenarnya dan kolom yang merepresentasikan label prediksi. Elemen diagonal dari matriks menunjukkan jumlah prediksi yang benar (TP dan TN), sedangkan elemen non-diagonal menunjukkan jumlah prediksi yang salah (FP dan FN). Confusion matrix digunakan untuk menghitung berbagai metrik klasifikasi seperti yang sudah dijelaskan pada $\textbf{soal 1}$. Berikut adalah contoh confusion matrix untuk model klasifikasi biner:
| Actual\Prediction | Prediction Positive | Prediction Negative |
|:--|:--|:--|
| Actual Positive | True Positive | False Negative |
| Actual Negative | False Positive | True Negative |

Referensi: https://medium.com/@pararawendy19/memahami-metrik-pada-pemodelan-klasifikasi-29cd5b738ee7

Selanjutnya, $\textit{Receiver-Operating Characteristic (ROC) Curve}$ merupakan kurva yang menggambarkan hubungan antara True Positive Rate (TPR) dan False Positive Rate (FPR) ketika ambang batas klasifikasi bervariasi. Pada ROC, setiap titik mewaliki satu ambang batas tertentu. $\textit{Area Under Curve (AUC)}$ merupakan metrik yang mengukur luas di bawah kurva ROC, yang berkisar dari 0 hingga 1 dan apabila nilai AUC semakin tinggi, maka semakin baik model klasifikasi tersebut.

Referensi:
https://medium.com/@andimrinaldisaputraa/memahami-dan-menerapkan-matriks-evaluasi-roc-auc-dalam-machine-learning-4468e5fcb9a

````python
# Import library
import keras, numpy as np, matplotlib.pyplot as plt, tensorflow as tf

NIM = 13219043
````

````python
# Load dataset
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

# Normalisasi data
train_images = train_images / 255.0
test_images = test_images / 255.0
````

````python
# Membuat model neural network dengan 3 hidden layer
model1_soal5 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation=tf.nn.sigmoid),
    tf.keras.layers.Dense(64, activation=tf.nn.sigmoid),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

# Compile model
model1_soal5.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train model
model1_soal5.fit(train_images, train_labels, epochs=10, batch_size=32, validation_data=(test_images, test_labels))
````

````python
# Membuat model neural network dengan 3 hidden layer dan 1 dropout layer
model2_soal5 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation=tf.nn.sigmoid),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(64, activation=tf.nn.sigmoid),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

# Compile model
model2_soal5.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train model
model2_soal5.fit(train_images, train_labels, epochs=10, batch_size=32, validation_data=(test_images, test_labels))
````

````python
# Membuat model neural network dengan 3 hidden layer, 1 dropout layer, dan 1 batch normalization layer
model3_soal5 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation=tf.nn.sigmoid),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(128, activation=tf.nn.sigmoid),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

# Compile model
model3_soal5.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train model
model3_soal5.fit(train_images, train_labels, epochs=10, batch_size=32, validation_data=(test_images, test_labels))
````

````python
# Membuat model neural network dengan 4 hidden layer dengan nilai sigmoid yang berbeda dan 1 dropout layer
model4_soal5 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation=tf.nn.sigmoid),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(128, activation=tf.nn.sigmoid),
    tf.keras.layers.Dense(64, activation=tf.nn.sigmoid),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

# Compile model
model4_soal5.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train model
model4_soal5.fit(train_images, train_labels, epochs=10, batch_size=32, validation_data=(test_images, test_labels))
````

````python
# Membuat model neural network dengan 3 hidden layer dengan aktivasi relu, 1 dropout layer, dan 1 batch normalization layer
model7_soal5 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

# Compile model
model7_soal5.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train model
model7_soal5.fit(train_images, train_labels, epochs=10, batch_size=32, validation_data=(test_images, test_labels))
````

## Jawaban Soal 9
### Singular Value Decomposition (SVD)

Singular Value Decomposition (SVD) adalah sebuah teknik faktorisasi matriks yang sangat penting dalam bidang komputasi numerik dan analisis data. Diberikan sebuah matriks $ X $ dengan ukuran $ m \times n $, SVD menguraikan $ X $ menjadi tiga matriks: $ U $, $ \Sigma $, dan $ V^* $, dimana:

- $U$ adalah matriks ortogonal $m \times m $.
- $\Sigma$ adalah matriks diagonal $m \times n $ dengan nilai-nilai singular di diagonalnya.
- $V^*$ (konjugat transpose dari $ V $) adalah matriks ortogonal $ n \times n $.

Formulasi SVD adalah:
$X = U \Sigma V^*$

Nilai-nilai pada $ \Sigma $ dikenal sebagai nilai-nilai singular dari $ X $ dan kolom-kolom $ U $ dan $ V $ dikenal sebagai vektor-vektor singular kiri dan kanan dari $ X $.

### Hubungan SVD dengan PCA (Principal Component Analysis)

PCA adalah teknik yang digunakan untuk reduksi dimensi dalam analisis data, sering digunakan untuk mengekstrak fitur penting dari kumpulan data berdimensi tinggi. Hubungan antara SVD dan PCA adalah sebagai berikut:

1. **Menggunakan SVD untuk PCA**: PCA dapat diimplementasikan melalui SVD. Dalam konteks PCA, kita biasanya bekerja dengan matriks data terpusat $ X $, yang artinya setiap kolom dari $ X $ telah dikurangi dengan rata-rata kolomnya.

2. **Ekstraksi Komponen Utama**: Setelah menghitung SVD dari matriks terpusat $ X $, vektor-vektor singular kanan ($ V $) dari SVD sesungguhnya merupakan komponen utama (principal components) dari $ X $. Ini karena $ V $ mengandung arah-arah di mana data memiliki varians maksimum.

3. **Transformasi Data**: Matriks $ \Sigma $ berisi nilai-nilai singular yang berkorelasi dengan seberapa penting atau beratnya setiap komponen utama. Ketika kita mengalikan $ U $, $ \Sigma $, dan $ V^* $, kita efektif melakukan transformasi data ke basis komponen utama, seringkali hanya mempertahankan beberapa komponen utama pertama (yaitu, beberapa kolom pertama dari $ U $ dan $ V $, dan nilai terbesar dari $ \Sigma $) untuk reduksi dimensi.

4. **Varians Terjelaskan**: Dalam PCA, varians yang dijelaskan oleh setiap komponen utama sesuai dengan kuadrat dari nilai-nilai singular di $ \Sigma $.

Melalui hubungan ini, SVD tidak hanya membantu dalam menghitung PCA tetapi juga memberikan pemahaman yang lebih mendalam tentang struktur data, seperti mengidentifikasi dimensi-dimensi data yang paling informatif.

## Jawaban Soal 10

Pendekatan aproksimasi dapat dituliskan dengan penggunaan rumus berikut.
$$
\hat{X} = \sum_{k=1}^{r} \sigma_k u_k v_k^* = \sigma_1 u_1 v_1^* + \sigma_2 u_2 v_2^* + \cdots + \sigma_r u_r v_r^*
$$

dimana $\hat{X}$ adalah aproksimasi rank-$\textit{r}$ dari matrix $\textbf{X}$, $\sigma_k$ adalah nilai singular ke-$\textit{k}$ dari $\textbf{X}$, $\textbf{u}_k$ dan $\textbf{v}_k$ adalah kolom ke-$\textit{k}$ dari matrix $\textbf{U}$ dan $\textbf{V}$.

````python
# Import library
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
````

````python
# Fungsi untuk mengubah gambar menjadi grayscale
def rgb_to_gray(rgb):
    return np.dot(rgb[..., :3], [0.299, 0.587, 0.114])
````

````python
plt.clf() # membersihkan gambar sebelumnya
plt.imshow(img_gray, cmap=plt.get_cmap('gray')) # menampilkan gambar grayscale
plt.xticks([]) # menghilangkan sumbu x
plt.yticks([]) # menghilangkan sumbu y
plt.title('Gambar Awal Grayscale')
````
